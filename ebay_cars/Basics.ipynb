## Exploring Ebay Car Sales Data
We will work with a dataset of used cars from eBay Kleinanzeigen, a classifieds section of the German eBay website scraped by Kaggle user [orgesleka](https://www.kaggle.com/orgesleka). The original file can be found [here](https://data.world/data-society/used-cars-data).
The original dataset was modified by the authors of the challenge:
- a sample of 50,000 data points was selected from the full dataset
- the dataset was "dirtified" a bit to more closely resemble what you would expect from a scraped dataset

The dataset includes following columns:
 - `dateCrawled` - When this ad was first crawled. All field-values are taken from this date.
 - `name` - Name of the car.
 - `seller` - Whether the seller is private or a dealer.
 - `offerType` - The type of listing
 - `price` - The price on the ad to sell the car.
 - `abtest` - Whether the listing is included in an A/B test.
 - `vehicleType` - The vehicle Type.
 - `yearOfRegistration` - The year in which the car was first registered.
 - `gearbox` - The transmission type.
 - `powerPS` - The power of the car in PS.
 - `model` - The car model name.
 - `kilometer` - How many kilometers the car has driven.
 - `monthOfRegistration` - The month in which the car was first registered.
 - `fuelType` - What type of fuel the car uses.
 - `brand` - The brand of the car.
 - `notRepairedDamage` - If the car has a damage which is not yet repaired.
 - `dateCreated` - The date on which the eBay listing was created.
 - `nrOfPictures` - The number of pictures in the ad.
 - `postalCode` - The postal code for the location of the vehicle.
 - `lastSeenOnline` - When the crawler saw this ad last online.
 
 The aim of this project is to **clean the data** and **analyze the included used car listings**.

## Cleaning columns with numeric values

As we have mentioned above apart from string value columns there are few columns presenting numeric information in the dataset. They could be valuable for our analysis:
- `price`
- `odometer`
- `powers_ps`

Last column is already formatted as int type.
Next step is to take a look at columns `price` and `odometer` to identify the way to format them.

```python
autos["price"].value_counts().tail()
```

```python
autos["odometer"].value_counts()
```

After taking look at these columns we found out what should be done next:
 - column `price` should be cleaned from currency signs, formatted to float with deleted "," thousand delimeters and renamed `"price_eur"`. There is also a large number of rows with 0 price, that also should be taken into consideration
 - column `odometer` should be cleaned from km signs, formatted to float with deleted "," thousand delimeters and renamed `"odometer_km"`

```python
#clean and convert values in price column
autos["price"] = autos["price"].str.replace("$", "", regex=False).str.replace(",", "").astype(int)
#clean and convert values in odometer column
autos["odometer"] = autos["odometer"].str.replace("km", "").str.replace(",", "").astype(int)
```

```python
#rename columns 
autos.rename({"price":"price_eur", "odometer":"odometer_km"}, axis=1, inplace=True)
```

```python
autos["price_eur"].describe()
```

```python
autos["odometer_km"].describe()
```

We have succesfully converted columns `price` and `odometer` to numeric values (both are float64 now) and renamed them to "price_eur" and "odometer_km"

Next step is to identify outliers in columns "price_eur" and "odometer_km".
We use `Series.unique().shape` to see how many unique values include these columns.

```python
#identify outliers in the column price_eur
print(autos["price_eur"].unique().shape)
print(autos["price_eur"].value_counts())
#find rows without outliers
autos = autos[autos["price_eur"].between(1,999989)]
print(autos["price_eur"].describe())
```

There were a number of entries with unrealistic prices which lay far above maximum values in `price_eur` column, so rows with price>=999990 and with price=0 were treated as outliers.
The same operation for `odometer_km` column doesn't show obvious outliers.

## From data exploration and cleaning to analysis
### Top brands and average prices and mileage

On the stage of dataset exloration and data cleaning we have already found top-6 brands sold on e-bay: 
 - Volkswagen
 - Opel
 - BMW
 - Mercedes-Benz
 - Audi
 - Ford
 
Next step is to examine price differences among these brands.

```python
#create empty dictionaries to store the results
mean_price_by_brand = {}
mean_mileage_by_brand = {}
#create a list with top-6 brands
top_6_brands = ["volkswagen", "opel","bmw", "mercedes_benz", "audi", "ford"]
#iterate over top-6 brands to collect mean prices and millage
for b in top_6_brands:
    selected_rows = autos[autos["brand"]==b]
    price = selected_rows["price_eur"].mean()
    mileage = selected_rows["odometer_km"].mean()
    mean_price_by_brand[b] = round(price, 2)
    mean_mileage_by_brand[b] = round(mileage, 2)
print(mean_price_by_brand)
print(mean_mileage_by_brand)
```

Brands with top mean prices looks not so popular, among top-sellers it's quite understandable why volkswagen is at the first place: mean price is 5231 euro, which looks like a middle between price group of Mercedes Benz, Audi and BMW on one side and Opel and Ford on the other.
Mean mileage fluctuates slighly: for BMW and Mercedes Benz it is a little higher.

How does mileage affects these average numbers?
Trying to answer that question we should compare mean price and mean mileage for every top-6 brand.

We need to present that difference more clearly, next step is to create a table with these average measures.

```python
#create dataframe with series mean_price and mean_mileage
price_brand_serie = pd.Series(mean_price_by_brand)
means_df = pd.DataFrame(price_brand_serie, columns=["mean_price"])

mileage_brand_serie = pd.Series(data=mean_mileage_by_brand)
means_df.assign(mean_mileage=mileage_brand_serie)
```
